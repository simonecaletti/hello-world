```python
df = pd.read_csv(r'db.csv')
print(df)
```

```python
occurrency = np.array(df.groupby("specie").size())
print(occurrency)
```

```python
df.groupby("specie").get_group("Hatterene")
```

```python
df.groupby("specie").specie.count()
```

```python
fig, ax = plt.subplots(1, 1, sharex='col', sharey='row', dpi=300)
usage = df.groupby("specie").size().plot(kind="bar", ax=ax).get_figure()
ax.grid(axis="y", linewidth=0.5)
ax.set_yticks(range(14))
ax.set_xlabel("")
ax.set_title("TOP USAGE")
plt.tight_layout()
usage.savefig("usage.pdf")
files.download("usage.pdf")
```
